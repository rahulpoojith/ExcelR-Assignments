## Now convert into a list formate 

```python
list=[]
for i in range(0,7501):
    list.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
```

```python
te_arry=te.fit(list).transform(list)
```

```python
df_new=pd.DataFrame(te_arry,columns=df.columns)
df_new
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_item=apriori(df_new,min_support=0.5,use_colnames=True)
frequent_item
```

```python
## So we need to decrease a support value beacuse only single item is cross the gvalue of support 
```

```python
frequent_item=apriori(df_new,min_support=0.2,use_colnames=True)
frequent_item
```

### Association rules

```python
association_rules(frequent_item,min_threshold=.4,metric='confidence')
```

```python
association_rules(frequent_item,min_threshold=0.2,metric='support')
```

```python
association_rules(frequent_item,min_threshold=.4,metric='lift')
```
