## 1. Data Preparation:

survived = train_df[train_df['Survived'] == 1]: This line creates a new DataFrame called survived containing only the rows from the original train_df where the 'Survived' column has a value of 1 (indicating passengers who survived).

not_survived = train_df[train_df['Survived'] == 0]: Similarly, creates a new DataFrame not_survived containing rows where 'Survived' is 0 (passengers who didn't survive).

2. Calculation and Printing:

len(survived): Calculates the number of rows in the survived DataFrame, representing the count of survivors.

len(not_survived): Calculates the number of rows in the not_survived DataFrame, representing the count of non-survivors.

len(train_df): Calculates the total number of rows in the original train_df, representing the total number of passengers.

float(len(survived))/len(train_df)*100.0: Calculates the percentage of survivors by dividing the number of survivors by the total number of passengers and multiplying by 100.

float(len(not_survived))/len(train_df)*100.0: Calculates the percentage of non-survivors similarly.

The print statements then display the calculated counts and percentages in a formatted manner.

train_df is a Pandas DataFrame containing information about passengers, including a 'Survived' column with values 0 (not survived) and 1 (survived).

The %.1f format specifier in the print statements ensures that the percentages are displayed with one decimal place.
This code provides a basic overview of the survival rate but doesn't offer any insights into the factors affecting survival. Further analysis would be required to explore relationships between survival and other variables in the dataset.

## Feature extraction

In this section, we select the appropriate features to train our classifier. Here, we create new features based on existing features. We also convert categorical features into numeric form.

```python
train_test_data = [train_df, test_df] # combining train and test dataset

for dataset in train_test_data:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.')
```

.extract(' ([A-Za-z]+)\.'):
    
([A-Za-z]+): Matches one or more consecutive uppercase or lowercase letters ([A-Za-z]) and captures them in a group (indicated by parentheses).
    
\s: Matches a whitespace character (space, tab, newline, etc.).
    
\.: Matches a literal dot (.).
    
    This regular expression effectively extracts the capitalized word followed by a dot, assuming it represents a title (e.g., "Mr.", "Mrs.", "Dr.")
    
    

```python
train_df.tail()
```

```python
pd.crosstab(train_df['Title'], train_df['Sex'])
```

```python
for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col', \
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Other')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    
train_df[['Title', 'Survived']].groupby(['Title'], as_index=False).mean()
```

## Age Feature

We first fill the NULL values of Age with a random number between (mean_age - std_age) and (mean_age + std_age).

We then create a new column named AgeBand. This categorizes age into 5 different age range.

```python
for dataset in train_test_data:
    age_avg = dataset['Age'].mean()
    age_std = dataset['Age'].std()
    age_null_count = dataset['Age'].isnull().sum()
    
    age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
    dataset['Age'][np.isnan(dataset['Age'])] = age_null_random_list
    dataset['Age'] = dataset['Age'].astype(int)
    
train_df['AgeBand'] = pd.cut(train_df['Age'], 5)

print (train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean())
```

Generating Random Ages:

age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count): Generates a random list of integers with the same length as the number of missing values. The values are drawn from a uniform distribution between age_avg - age_std and age_avg + age_std. This approach aims to replace missing ages with values that are within a reasonable range based on the observed age distribution.

Filling Missing Values:

dataset['Age'][np.isnan(dataset['Age'])] = age_null_random_list: Assigns the generated random values to the missing values in the Age column.
dataset['Age'] = dataset['Age'].astype(int): Converts the Age column to integer data type, likely for subsequent calculations or analysis.

Creating Age Bands:

train_df['AgeBand'] = pd.cut(train_df['Age'], 5): Divides the Age column in the train_df DataFrame into 5 equal-sized bins (or "bands") and assigns a category to each row based on the bin it falls into.

Grouping and Calculating Mean:

train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean() groups the train_df DataFrame by the AgeBand column and calculates the mean of the Survived column for each age band.
as_index=False ensures that the grouped DataFrame is not converted to an index, making it easier to work with.

## let the convert ageband into numeric values using label encoding

```python
for dataset in train_test_data:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] = 4
```

### fare feature

## Replace missing Fare values with the median of Fare

```python
for dataset in train_test_data:
    dataset['Fare'] = dataset['Fare'].fillna(train_df['Fare'].median())
```

```python
train_df.head()
```

## lets fare divide into 4 categories

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 4)
print (train_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean())
```

pd.qcut divides the Fare column into 4 quantiles (equal-sized groups) based on their values. 
This means that each group will contain approximately the same number of passengers.

## those who have paid high amount have highest chances of survival

## now encode fareband into different categories

```python
for dataset in train_test_data:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
```
