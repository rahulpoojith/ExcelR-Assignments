### What techniques can be used to address multicollinearity in multiple linear regression?

#####  Multicollinearity occurs when two or more predictor (independent) variables in a multiple linear regression model are highly correlated. 
#### High multicollinearity can cause several issues in regression models, including unstable coefficient estimates, inflated standard errors, and difficulty in determining the effect of individual predictors. 

#### Here are several techniques that can be used to address multicollinearity in multiple linear regression: 

#### 1.Regularization Techniques:
Ridge Regression: Adds a penalty term to the regression equation based on the square of the coefficients. This can help reduce the impact of multicollinearity.

Lasso Regression: Adds a penalty term based on the absolute value of the coefficients. This can help shrink coefficients toward zero and reduce the effect of multicollinearity

. Elastic Net: A combination of Lasso and Ridge regression that can be useful when there is high multicollinearity.

#### 2.Transformation of Variables:
Sometimes transforming the predictor variables (e.g., log transformation) can help address multicollinearity.
