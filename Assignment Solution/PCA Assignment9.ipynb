Task 2: Dimensionality Reduction with PCA:
        
1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.

2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.

3.	Transform the original dataset into the principal components.


```python
from sklearn.preprocessing import StandardScaler
std_sca= StandardScaler()
```

```python
features= std_sca.fit_transform(df)
```

```python
features
```

```python
# Convert the standardized data back to a DataFrame
features = pd.DataFrame(features)
```

```python
features
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
```

```python
# Fit PCA on the standardized data
pca.fit(features)
```

```python
# Transform the data
df_pca = pca.transform(features)
```

```python
df_pca
```

```python
# Plot the explained variance ratio
plt.figure(figsize=(10, 8))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Explained Variance Ratio by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid()
plt.show()
```

```python
# Compute cumulative explained variance
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

```python
# Define the number of components to keep
n_components = 14  # Example: choosing 14 components

# Initialize PCA with the chosen number of components
pca = PCA(n_components=n_components)

# Fit and transform the data
df_pca_reduced = pca.fit_transform(features)

# Convert to DataFrame for easier handling
df_pca_reduced = pd.DataFrame(df_pca_reduced, columns=[f'PC{i+1}' for i in range(n_components)])

# Display the reduced data
print(df_pca_reduced.head())
```

```python
pca.explained_variance_ratio_
```

Standardize Features: Scaling the features so they have mean 0 and variance 1.
    
Apply PCA: Reduce dimensionality while retaining variance.
    
Explained Variance Ratio: Evaluate how much variance each principal component explains.
    
Cumulative Explained Variance: Decide on the number of components to retain based on the cumulative variance plot.
    
Transform Data: Apply PCA to get the reduced feature set

Task 3: Clustering with Original Data:
        
1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.

2.	Visualize the clustering results using appropriate plots.

3.	Evaluate the clustering performance using metrics such as silhouette score or Daviesâ€“Bouldin index.


```python
from sklearn.cluster import KMeans
```

```python
# Determine the optimal number of clusters using the elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=100)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid()
plt.show()
```

```python
# Set the optimal number of clusters (e.g., 3 from the elbow method)
n_clusters = 3

# Initialize and fit K-means
kmeans = KMeans(n_clusters=n_clusters, random_state=100)
kmeans.fit(features)

# Get cluster labels
df['Cluster'] = kmeans.labels_

# Display the first few rows with cluster labels
print(df.head())
```

Prepare the Data: Standardize the dataset to ensure consistency in feature scales.
    
Apply K-Means Clustering: Use the elbow method to determine the optimal number of clusters and then fit the K-means algorithm.
    
Analyze and Interpret: Examine cluster centers and visualize the clusters to interpret the results.

```python
# Apply PCA to reduce to 2 dimensions for visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(features)

# Create a DataFrame with PCA results and cluster labels
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_pca['Cluster'] = df['Cluster']

# Plot the clusters in 2D space
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Cluster', palette='viridis', marker='o')
plt.title('Clusters in 2D PCA Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Add cluster labels to the original DataFrame
df_with_clusters = pd.DataFrame(features, columns=df.columns[:-1])
df_with_clusters['Cluster'] = df['Cluster']
```

```python
# Plot cluster centers in 2D PCA space
df_cluster_centers_pca = pca.transform(kmeans.cluster_centers_)

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='Cluster', palette='viridis', marker='o', alpha=0.7)
sns.scatterplot(x=df_cluster_centers_pca[:, 0], y=df_cluster_centers_pca[:, 1], color='red', s=200, marker='X', label='Centroids')
plt.title('Clusters with Centroids in 2D PCA Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Compute the silhouette score
silhouette_avg = silhouette_score(features, kmeans.labels_)
print(f'Silhouette Score: {silhouette_avg:.2f}')



# Silhouette Score
#The Silhouette Score measures how similar each point is to its own cluster compared to other clusters. It ranges from -1 to 1, where:

 #1 indicates that the sample is well-clustered.
 #0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters.
#-1 indicates that the sample is incorrectly clustered.
```

```python
# Compute evaluation metrics
silhouette_avg = silhouette_score(features, kmeans.labels_)
print(f'Silhouette Score: {silhouette_avg:.2f}')
```

Discuss Similarities or Differences in Clustering Results

Similarities:

Cluster Count: Both the PCA-transformed data and original data use the same number of clusters (3), so the number of identified clusters will be consistent.

Overall Trends: In some cases, the overall trend or distribution of clusters might appear similar if PCA has captured significant variance that aligns with the original feature space.

Differences:

Cluster Separation: PCA often provides a clearer separation of clusters in 2D space. This is due to PCA focusing on the principal components that capture the maximum variance, potentially leading to more distinct clusters.

Cluster Shape and Size: The shape and distribution of clusters might be different in PCA space compared to the original space. PCA can project data in a way that enhances separation, which might not always directly map to the original feature space.

Interpretability: Clusters in PCA space might not be directly interpretable in terms of original features. The reduced-dimensional view might mask some details present in the original high-dimensional space.

3. Reflect on the Impact of Dimensionality Reduction on Clustering Performance

Positive Impacts:

Enhanced Separation: PCA can enhance clustering performance by making the clusters more distinct due to the focus on principal components with maximum variance.

Reduced Noise: By removing less important features, PCA reduces the impact of noise and irrelevant data, leading to potentially more meaningful clusters.

Improved Visualization: Dimensionality reduction facilitates visualization, making it easier to see the structure and separation of clusters.

Negative Impacts:

Information Loss: Dimensionality reduction may discard some important information, which could affect clustering accuracy if the discarded dimensions had meaningful variance.

Interpretability Challenges: Clusters in the PCA space are defined by principal components, which might be harder to interpret in terms of the original features.

4. Analyze the Trade-Offs Between Using PCA and Clustering Directly on the Original Dataset

Advantages of PCA:

Simplified Clustering: PCA reduces the data to a lower-dimensional space, which simplifies the clustering process and can enhance cluster separation.

Noise Reduction: PCA filters out noise and irrelevant features, potentially leading to more stable and meaningful clusters.

Visualization: PCA makes it easier to visualize clustering results, especially in 2D or 3D plots.

Disadvantages of PCA:

Loss of Information: Important details might be lost during dimensionality reduction, which can affect the quality of clustering if the principal components do not fully capture the data's structure.

Reduced Interpretability: Clusters are defined in terms of principal components rather than original features, making interpretation and practical implications less straightforward.

Advantages of Original Data:

Full Utilization: Using the original dataset means all available features are utilized, preserving all available information for clustering.

Direct Interpretation: Clusters are directly related to the original features, making it easier to understand and relate the results to real-world variables

Disadvantages of Original Data:

High Dimensionality: Clustering in a high-dimensional space can be challenging due to the curse of dimensionality, where distance metrics become less meaningful and clusters may overlap.

Inclusion of Noise: Original data may contain noise and redundant features that can negatively impact clustering performance.
Summary






## Conclusion


Key Findings: PCA often enhances clustering by improving separation and reducing noise, but may also result in some loss of information and interpretability. Clustering directly on original data preserves feature richness but can be affected by high dimensionality and noise.

Practical Implications: PCA helps in handling high-dimensional data, simplifying analysis and visualization, while clustering directly on the original data provides a fuller picture of the feature space.

Recommendations: Use PCA when dealing with high-dimensional data or when visualization and noise reduction are priorities. Clustering directly on original data is suitable when the dataset is not excessively large and preserving all features is essential.
